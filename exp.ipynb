Looks like data is clean beacuse we don't have any NaN

```python
#dropping unneccessary column
data = data.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)
data
```

```python
data['Likes'].fillna(data['Likes'].mean(), inplace=True)
data
```

```python
data.dropna(inplace=True)
```

We need to check the outliners

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#plotting the data to check the outliners
sns.boxplot(data['Likes'])
```

There are no outliners, but if we have extream outliners then we have to remove them

```python
Q1 = data['Likes'].quantile(0.25)
Q3 = data['Likes'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

data = data[(data['Likes'] >= lower_bound) & (data['Likes'] <= upper_bound)]
```

```python
#plotting the data to check the outliners
sns.boxplot(data['Likes'])
```

```python
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
```

```python
# Check for duplicates
print(data.duplicated().sum())

# Drop duplicates
data = data.drop_duplicates()
```
